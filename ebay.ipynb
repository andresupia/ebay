```python
# Instalación de librerías necesarias
!pip install selenium pandas webdriver-manager openpyxl

# Importar las librerías necesarias
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Configurar el WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# Navegar al sitio web de eBay
driver.get('https://www.ebay.com/sch/i.html?_nkw=products')

# Asegurarnos de que la página haya cargado completamente
time.sleep(5)

# Extraer los datos de los productos
products = []
product_elements = driver.find_elements(By.CSS_SELECTOR, 'li.s-item')

for product in product_elements:
    try:
        name = product.find_element(By.CSS_SELECTOR, 'span[role="heading"], h3.s-item__title').text
        if name == '':
            name = 'N/A'
    except:
        name = 'N/A'
    try:
        price = product.find_element(By.CSS_SELECTOR, 'span.s-item__price').text
    except:
        price = 'N/A'
    try:
        shipping = product.find_element(By.CSS_SELECTOR, 'span.s-item__logisticsCost').text
    except:
        shipping = 'N/A'
    try:
        # Ajustar para incluir porcentaje de descuento si es aplicable
        discount = product.find_element(By.CSS_SELECTOR, 'span.s-item__discount').text
    except:
        discount = 'N/A'
    
    products.append({
        'name': name,
        'price': price,
        'shipping': shipping,
        'discount': discount
    })

# Verificar si los productos fueron extraídos correctamente
for product in products:
    print(product)

# Convertir la lista de productos a un DataFrame y guardarlo en un archivo Excel
df = pd.DataFrame(products)
df.to_excel('ebay_products.xlsx', index=False)

# Cerrar el navegador
driver.quit()
```
